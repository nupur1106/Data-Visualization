# Google Playstore Case Study

**Problem Statement**

The team at Google Play Store wants to develop a feature that would enable them to boost visibility for the most promising apps. Questions like:
- Does a higher size or price necessarily mean that an app would perform better than the other apps? 
- Or does a higher number of installs give a clear picture of which app would have a better rating than others?


```python
#import the libraries
import pandas as pd
import numpy as np
```

```python
#the dataset and  the first five rows
inp0 = pd.read_csv("googleplaystore_v2.csv")
inp0.head()
```

```python
#shape of the dataframe
inp0.shape
```

### Data Handling and Cleaning

The first few steps involve making sure that there are no __missing values__ or __incorrect data types__  These  problems are handled as follows:

 - For Missing Values: Some common techniques to treat this issue are
    - Dropping the rows containing the missing values
    - Imputing the missing values
    - Keep the missing values if they don't affect the analysis
 
    
 - Incorrect Data Types:
    - Clean certain values 
    - Clean and convert an entire column
 

```python
#datatypes of all the columns of the dataframe
inp0.info()
```

#### Missing Value Treatment

```python
#number of null values in the columns
inp0.isnull().sum()
```

Handling missing values for rating
 - Ratings is the target variable
 - drop the records

```python
#Droping the rows having null values in the Rating field
inp1 = inp0[~inp0.Rating.isnull()]

# shape of the dataframe
inp1.shape
```

```python
#  number of nulls in the Rating field again to cross-verify
inp1.Rating.isnull().sum()
```

```python
#number of nulls in the dataframe again and total number of null values
inp1.isnull().sum()
```

```python
#nulls in the Android Version column
inp1[inp1['Android Ver'].isnull()]
```

```python
#Droping the row having shifted values
inp1 = inp1[~(inp1['Android Ver'].isnull() & (inp1.Category == "1.9"))]
#nulls again in Android version column to cross-verify
inp1[inp1['Android Ver'].isnull()]
```

Imputing Missing Values

- For numerical variables use mean and median
- For categorical variables use mode

```python
#the most common value in the Android version column
inp1['Android Ver'].value_counts()
```

```python
#Filling the nulls in the Android Version column with the above value
inp1['Android Ver'] = inp1['Android Ver'].fillna(inp1['Android Ver'].mode()[0])
```

```python
#the nulls in the Android version column again to cross-verify
inp1['Android Ver'].isnull().sum()
```

```python
#the nulls in the entire dataframe again
inp1.isnull().sum()
```

```python
#the most common value in the Current version column
inp1['Current Ver'].value_counts()
```

```python
#Replacing the nulls in the Current version column with the above value
inp1['Current Ver'] = inp1['Current Ver'].fillna(inp1['Current Ver'].mode()[0])
```

```python
# the most common value in the Current version column again
inp1['Current Ver'].value_counts()
```

#### Handling Incorrect Data Types 


```python
#the datatypes of all the columns 
inp1.dtypes
```

```python
#calculating the average price of all apps having the Android version as "4.1 and up" 
inp1.head()
```

```python
#the Price column to check the issue
inp1.Price.value_counts()
```

```python
# function to make the changes
inp1.Price = inp1.Price.apply(lambda x: 0 if x=="0" else float(x[1:]))
```

```python
#the dtype of Price once again
inp1.Price.dtype
```

```python
#Analyzing the Reviews column
inp1.Reviews.value_counts()
```

```python
#Changing the dtype of this column
inp1.Reviews = inp1.Reviews.astype("int32")

inp1.Reviews.describe()
```

```python
#Analyzing the Installs Column
inp1.Installs.head()
```

```python
# approximate number of apps at the 50th percentile.
def clean_installs(val):
    return int(val.replace(",","").replace("+",""))
type(clean_installs("3,000+"))
inp1.Installs = inp1.Installs.apply(clean_installs)
inp1.Installs.describe()
```

#### Sanity Checks

The data that we have needs to make sense and so we can perform certain sanity checks on them to ensure they are factually correct . Some sanity checks can be:

- Rating is between 1 and 5 for all the apps.
- Number of Reviews is less than or equal to the number of Installs.
- Free Apps shouldn’t have a price greater than 0.


```python
# the sanity checks on the Reviews column
inp1[(inp1.Reviews > inp1.Installs)].shape
```

```python
inp1[(inp1.Reviews > inp1.Installs)]
```

```python
inp1 = inp1[inp1.Reviews <= inp1.Installs]
```

```python
#the sanity checks on prices of free apps 
inp1[(inp1.Type == "Free") & (inp1.Price>0)]
```

#### Heat Maps

```python
##Ratings vs Size vs Content Rating

##using buckets for the Size column using pd.qcut

inp1['Size_Bucket'] = pd.qcut(inp1.Size, [0, 0.2, 0.4, 0.6, 0.8, 1], ["VL","L","M","H","VH"])
```

```python
## a pivot table for Size_buckets and Content Rating with values set to Rating
inp1.head()
pd.pivot_table(data=inp1, index="Content Rating", columns="Size_Bucket", values="Rating")
```

```python
##the aggregation to median
pd.pivot_table(data=inp1, index="Content Rating", columns="Size_Bucket", values="Rating", aggfunc=np.median)
```

```python
##the aggregation to 20th percentile
pd.pivot_table(data=inp1,index="Content Rating",columns="Size_Bucket",values="Rating",aggfunc=lambda x: np.quantile(x,0.2))
```

```python
##Storing the pivot table in a separate variable
res = pd.pivot_table(data=inp1,index="Content Rating",columns="Size_Bucket",values="Rating",aggfunc=lambda x: np.quantile(x,0.2))
```

```python
## a heat map
sns.heatmap(res)
plt.show()
```

```python
##Applying customisations
sns.heatmap(res, cmap = "Greens", annot=True)
plt.show()
```

#### Stacked Bar Charts to breaks down each bar of the bar chart on the basis of a different category

```python
## Creating a pivot table for Content Rating and updateing Month with the values set to Installs
pd.pivot_table(data=inp1, values="Installs", index="updated_month", columns="Content Rating", aggfunc=sum)
```

```python
##Storing the table in a separate variable
monthly = pd.pivot_table(data=inp1, values="Installs", index="updated_month", columns="Content Rating", aggfunc=sum)
```

```python
## the stacked bar chart.
monthly.plot(kind="bar", stacked="True", figsize=[10,6])
plt.show()
```

```python
## the stacked bar chart again wrt to the proportions.
monthly_perc = monthly[["Everyone","Everyone 10+","Mature 17+","Teen"]].apply(lambda x: x/x.sum(), axis=1)
```
